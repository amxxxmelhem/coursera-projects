```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

# Step 1: Load the webpage
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")

# Step 2: Find all tables
tables = soup.find_all("table")

# Step 3: Loop through to find the right one
revenue_table = None
for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        revenue_table = table
        break

# Step 4: Error check
if revenue_table is None:
    print("❌ Revenue table not found")
else:
    # Step 5: Use StringIO to safely read the table
    table_html = str(revenue_table)
    tesla_revenue = pd.read_html(StringIO(table_html))[0]

    # Step 6: Clean the data
    tesla_revenue.columns = ['Date', 'Revenue']
    tesla_revenue['Revenue'] = tesla_revenue['Revenue'].str.replace('$', '').str.replace(',', '')
    tesla_revenue.dropna(inplace=True)

    # Step 7: Show last rows
    print(tesla_revenue.tail())
```
